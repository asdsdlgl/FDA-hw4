```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets, linear_model
from sklearn.svm import SVC
from sklearn.neural_network import MLPClassifier
```

```python
attribute = ['Win', 'Cluster ID', 'Game mode', 'Game type']
attribute += ['Hero%d' %(i+1) for i in range(113) ]
#attribute += [i for i in range(113) ]
train = pd.read_csv("dota2Train.csv",names=attribute)
test = pd.read_csv("dota2Test.csv",names=attribute)
train.head(5)
```

```python
y_train = train['Win']
x_train = train.loc[:, ['Hero%d' %(i+1) for i in range(113)]]
#x_train = train[['Hero%d' %(i+1) for i in range(113)]]
x_train['Game mode'] = train['Game mode']
x_train['Game type'] = train['Game type']

y_test = test['Win']
x_test = test.loc[:, ['Hero%d' %(i+1) for i in range(113)]]
#x_test = train[['Hero%d' %(i+1) for i in range(113)]]
x_test['Game mode'] = test['Game mode']
x_test['Game type'] = test['Game type']

x_train.head()
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn import metrics

clf = LogisticRegression(fit_intercept=True,solver='lbfgs',C=1e15)
Regression = clf.fit(x_train, y_train)
Reg_result = Regression.predict(x_test)
accuracy = metrics.accuracy_score(Reg_result, y_test)
accuracy
```

```python
attr = ['Hero%d' %(i+1) for i in range(113) ]
q1 = 0
q2 = 0
q3 = 0
q4 = 0
number = []
for ele in attr :
    df_filter = train.loc[train[ele] != 0]
    df_filter[ele].value_counts()
    df_sort = df_filter.groupby(['Win', ele])
    if df_filter[ele].count() == 0:
        number.append(0)
        continue
    if 1 not in df_sort.size()[-1]:
        q1 = 0
    else:
        q1 = df_sort.size()[-1][1]
        
    if -1 not in df_sort.size()[-1]:
        q2 = 0
    else:
        q2 = df_sort.size()[-1][-1]
        
    if 1 not in df_sort.size()[1]:
        q3 = 0
    else:
        q3 = df_sort.size()[1][1]
        
    if -1 not in df_sort.size()[1]:
        q4 = 0
    else:
        q4 = df_sort.size()[1][-1]
        
    win = q2 + q3
    lose = q1 + q4
    if lose+win == 0:
        number.append(0)
    else:
        number.append(win/(lose+win))
number
```

```python
hero_x = range(0, 113)
# "bo" is for "blue dot"
plt.plot(hero_x, number, 'b', label='win rate')
plt.xlabel('hero')
plt.ylabel('win rate')
plt.legend()
```

```python
temp = x_train.copy()
for i in range(len(x_train.columns)-2):
    temp[temp.columns[i]] = number[i] * x_train[x_train.columns[i]]
```

```python
for index, row in temp.iterrows():
    temp.loc[index, 'ans'] = row[temp.columns[:113]].sum()
```

```python
temp['test'] = ((temp['ans']>=0).astype(int))*2-1
temp
```

```python
clf2 = LogisticRegression(fit_intercept=True,solver='lbfgs',C=1e15,max_iter=4000)
y_train2 = temp['test']
Regression2 = clf2.fit(x_train, y_train2)
Reg_result2 = Regression2.predict(x_test)
accuracy = metrics.accuracy_score(Reg_result2, y_test)
accuracy
```
